```python
def features_visualizer(df):
    plt.subplot()
    for column in df: 
        if is_numeric_dtype(df[column]) and column not in ["Is_Superhost", "Is_Exact_Location", "Instant_Bookable", "Business_Travel_Ready"]:
            #sns.displot(df[column], kde=True)
            sns.boxplot(df[column])
            #plt.savefig(f'Figures/{column}.png')
```

```python
features_visualizer(df)
```
